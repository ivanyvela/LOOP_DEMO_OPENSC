```python
import os
import pandas as pd
import re

# Define the root directory (current folder)
root_dir = os.getcwd()
index_file = 'Redox_Index.csv'

# Folders to process
folders = ['LOOP2', 'LOOP3', 'LOOP4', 'LOOP6', 'DEMO']

data = []

def extract_id(filename):
    # Check for "Punkt " or "Point " (case insensitive)
    match = re.match(r'^(Punkt|Point)\s+(.+?)_', filename, re.IGNORECASE)
    if match:
        return match.group(2)
    
    # Check for "P" followed immediately by the ID (e.g., P1a)
    # This specifically looks for 'P' at the start NOT followed by 'unkt' or 'oint'
    match_p = re.match(r'^P(?![unkt|oint])(.+?)_', filename, re.IGNORECASE)
    if match_p:
        return match_p.group(1)
    
    # Default: Grab everything from the start up to the first underscore
    return filename.split('_')[0]

# Loop through the folders
for folder in folders:
    folder_path = os.path.join(root_dir, folder)
    
    if not os.path.exists(folder_path):
        continue

    # Get all files and sort them to keep things tidy
    files = sorted(os.listdir(folder_path))

    for filename in files:
        # Match REMax, RE1Max, RE2Max, or RE3Max
        if "REMax" in filename or any(f"RE{i}Max" in filename for i in range(1, 4)):
            
            # Extract the ID using our logic
            file_id = extract_id(filename)
            
            # Create the Source_File string using backslashes for the CSV format
            source_path = f"{folder}\\{filename}"
            
            data.append({
                "Source_File": source_path,
                "ID": file_id,
                "Grounding": "",
                "Cal start": "",
                "Cal end": "",
                "Notes": "",
                "Interface_Human": "", #To be manually inputted
                "Second_Interface_Human": "", #To be manually inputted
                "Interface_Machine": "" # To be computed

            })

# Create DataFrame and save
df = pd.DataFrame(data)
df.to_csv(index_file, index=False)

print(f"Successfully processed {len(df)} files.")
```
